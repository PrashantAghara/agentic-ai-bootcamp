```python
from pydantic import BaseModel
```

```python
class Person(BaseModel):
    name: str
    age: int
    city: str

person = Person(name="Prashant", age=25, city="Nadiad")

print(person)
```

```python
type(person)
```

```python
## Data validation
person1 = Person(name="Prashant", age=25, city=12)
```

### Optional Fields

```python
from typing import Optional

class Employee(BaseModel):
    id: int
    name: str
    department: str
    salary: Optional[float] = None
    is_active: Optional[bool] = True
```

```python
emp1 = Employee(id=1, name="John", department="Engg")
print(emp1)
```

```python
emp2 = Employee(id=1, name="John", department="Engg", salary=1000)
print(emp2)
```

### Lists

```python
from typing import List

class Classroom(BaseModel):
    room_no: int
    students: List[str]
    capacity: int
```

```python
classroom = Classroom(
    room_no=1,
    students=["A", "B", "C"],
    capacity=50
)

classroom
```

### Nested Models

```python
class Address(BaseModel):
    street: str
    city: str
    zip_code: str

class Customer(BaseModel):
    cust_id: int
    name: str
    address: Address

cust1 = Customer(
    cust_id=1,
    name="Prashant",
    address=Address(
        street="1A",
        city="Nadiad",
        zip_code="387001"
    )
)

cust1
```

```python
from pydantic import Field

class Item(BaseModel):
    name: str = Field(min_length=2, max_length=50)
    price: float = Field(gt=0, le=1000)
    quantity: int = Field(gt=0)

item1 = Item(
    name="Iphone",
    price=100,
    quantity=1
)
item1
```

```python
item2 = Item(
    name="ABC",
    price=100,
    quantity=0
)
```

```python
class User(BaseModel):
    username: str = Field(..., description="Unique username for the user")
    age: int = Field(default=18, description="User's age, defualts to 18")
    email: str = Field(default_factory=lambda: "user@example.com", description="Default email address")

user1 = User(username="ABC")
print(user1)

user2 = User(username="PQE", age=20, email="pqr@gmail.com")
print(user2)
```
